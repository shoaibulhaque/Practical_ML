9.*Demean each column of Data2. Hint: use broadcasting.*

````python
data2 = np.array([[1, 5, 2, 0], [3, 6, 4, 7]])

# Calculate the mean of each column
mean_of_columns = np.mean(data2, axis=0)
print(mean_of_columns)
# Demean each column by subtracting the mean of the column
demeaned_data2 = data2 - mean_of_columns

print(demeaned_data2)
````

**Explanation:**

We calculate the mean of each column by taking the mean along axis 0 (rows) using `np.mean(data2, axis=0)`.

To demean each column, we subtract the calculated mean from the corresponding column using broadcasting `(data2 - mean_of_columns)`. Broadcasting ensures element-wise subtraction between the arrays.

````python
# Converting the data into arrays
student_1 = np.array([60,40,70])
student_2 = np.array([70,60,55])
student_3 = np.array([50,80,65])

# subjects
subjects_cred = np.array([2,3,4])
````

````python
def get_grade_point(marks):
    grade_points_cond = [
        marks >= 94,
        np.logical_and(marks >= 85, marks <= 93),
        np.logical_and(marks >= 80, marks <= 84),
        np.logical_and(marks >= 75, marks <= 79),
        np.logical_and(marks >= 70, marks <= 74),
        np.logical_and(marks >= 67, marks <= 69),
        np.logical_and(marks >= 64, marks <= 66),
        np.logical_and(marks >= 60, marks <= 63),
        np.logical_and(marks >= 57, marks <= 59),
        np.logical_and(marks >= 54, marks <= 56),
        np.logical_and(marks >= 50, marks <= 53),
        marks < 50
    ]
    grade_points = np.select(grade_points_cond, [4.0, 4.0, 3.7, 3.4, 3.0, 2.7, 2.4, 2.0, 1.7, 1.4, 1.0, 0.0])

    return grade_points
````

````python
def calc_gpa(subjects_cred, student_marks):
  total_cred = np.sum(subjects_cred)

  if total_cred > 0:
    GPA = np.sum(subjects_cred * get_grade_point(student_marks))/total_cred

  return GPA
````

````python
print(calc_gpa(subjects_cred, student_3))
````

14. *Compute the dot product and inverse of 4x4 random matrix. Hint: Use numpy.linalg functions.*

````python
# Generate a random 4x4 matrix
matrix = np.random.rand(4, 4)

# Calculate the dot product of the matrix with itself
dot_product = np.dot(matrix, matrix)

# Try-except block to handle potential singular matrix errors
try:
  # Calculate the inverse of the matrix
  inverse = np.linalg.inv(matrix)
  print("Dot product:\n", dot_product)
  print("Inverse:\n", inverse)
except np.linalg.LinAlgError as e:
  print("Error: The matrix may be singular and cannot be inverted.")
````
